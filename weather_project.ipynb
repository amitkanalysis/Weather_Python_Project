```python
plt.figure(figsize=(15, 5))
sns.countplot(x='Weather Condition', data=df, palette='viridis')
plt.title('Weather Condition Count')
plt.xticks(rotation= 90)
plt.show()
```

```python
df.head(2)
```

```python
plt.figure(figsize=(15, 10))
for col in ['Weather Condition', 'Visibility_km']:
    sns.lineplot(x='Wind Speed_km/h', y=col, data=df, label=col)
plt.title('Weather Condition and Visibility by Wind Speed ')
plt.xlabel('Weather Condition')
plt.ylabel('Value')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.histplot(df['Temp_C'], bins=30, color='skyblue',
             label='Temp_C', kde=True)
sns.histplot(df['Dew Point Temp_C'], bins=30,
             color='orange', label='Dew Point Temp_C', kde=True)
plt.title('Histogram of Temperature and Dew Point Temperature')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```
